## Divorce and marriage by age group
- ONS overview [page](https://www.ons.gov.uk/peoplepopulationandcommunity/birthsdeathsandmarriages/divorce/bulletins/divorcesinenglandandwales/2020) for divorce statistics
- [ONS data set on divorce](https://www.ons.gov.uk/file?uri=/peoplepopulationandcommunity/birthsdeathsandmarriages/divorce/datasets/divorcesinenglandandwales/2020/divorcesworkbook2020final31012022144352.xlsx)

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%load_ext lab_black
```

```python
# Read data
xls = pd.ExcelFile("./data/divorcesworkbook2020final31012022144352.xlsx")
# xls.sheet_names
df_divorce_by_year = pd.read_excel(xls, "8a")
```

```python
df_divorce_by_year.columns = df_divorce_by_year.iloc[4]
df_divorce_by_year = df_divorce_by_year.drop([0, 1, 2, 3, 4, 61])
for col in df_divorce_by_year.columns:
    df_divorce_by_year[col] = pd.to_numeric(df_divorce_by_year[col])
```

```python
df_divorce_by_year.head()
```

```python
df_to_plot = df_divorce_by_year.drop("Number of marriages", axis=1).set_index(
    "Year of marriage"
)
df_to_plot.columns.str.strip("Anniversary\n(Years)\n")
df_to_plot.columns = df_to_plot.columns.str.strip("Anniversary\n(Years)\n")
```

```python
df_to_plot
```

```python
# Plot
mask = df_to_plot.isnull()
cmap = sns.cubehelix_palette(
    as_cmap=True,
    light=0.9,
    dark=0.1,
)  # reverse=True)
sns.set(rc={"figure.figsize": (20, 12)})
sns.set_context("poster", font_scale=0.8)
ax = sns.heatmap(
    df_to_plot,
    mask=mask,
    # center=25,
    # cmap="coolwarm",
    cmap=cmap,
    linewidth=0.5,
    linecolor="w",
    cbar_kws={"label": "% divorced"},
)
ax.set_facecolor("white")
ax.set_xlabel("Years married")
ax.set_title("Percent of marriages divorced by duration of marriage (UK)", fontsize=27)
ax.annotate(
    "Married in 1967,\n32% divorced\nby 53th anniversary",
    color="grey",
    xy=(52.5, 4.5),
    xytext=(49, 13),
    arrowprops=dict(
        arrowstyle="simple",
        color="hotpink",
    ),
)
ax.annotate(
    "44% of couples married in 1990 were\ndivorced by their 30th anniversary",
    color="grey",
    xy=(29.3, 27.5),
    xytext=(31, 33),
    arrowprops=dict(
        arrowstyle="simple",
        color="hotpink",
    ),
)
ax.annotate(
    "Married in 2013,\n10% divorced\nby 7th anniversary",
    color="grey",
    xy=(6.5, 50.5),
    xytext=(13, 53),
    arrowprops=dict(
        arrowstyle="simple",
        color="hotpink",
    ),
)
ax.text(
    x=0.15,
    y=60,
    s="Source: ONS & data.meets.poetry",
    fontsize=13,
    fontname="Arial",
    color="#969696",
    ha="center",
    # transform=fig.transFigure,
)
plt.show()
```
